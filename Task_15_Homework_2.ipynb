### Задача_1 (решено)

Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:
- среднюю букву, если число букв в слове нечетное;
- две средних буквы, если число букв четное.

```python
#импортирование_системных_библиотек
import sys
import os

#библиотека_по_работе_с_датой_и_временем
import time

#библиотека_для_регулярных_выражений
import re
```

```python
%%time

#создание_функции_проверки_символа_и_вывода_на_печать
def check_simbol():
    #печать_приглашения_на_ввод    
    print("\nВведите слово на латиннице без пробелов:")
    
    #инициализация_переменной_и_функция_ввода
    strl = input()

    #формирование_словаря
    diction = "^[a-zA-Zа]+$"

    #создание_шаблона_для_проверки
    pattern = re.compile(diction)

    #проверка_на_символы
    check = (pattern.search(strl) is not None)

    #если_все_символы_латинские_то_работаем_дальше
    if(check == True):
      #печать_введенного_слова
      print("Вы ввели: ", strl)

      #разделение_строки_на_символы_и_запись_в_массив
      sMass = list(strl)
  
      #печать_массива
      print("Полученный символьный массив: ", sMass)
    
      #измерение_длины_массива
      sMassLen = len(sMass) 
  
      #если_остаток_от_деления_равен_0_значит_чет
      sHalfMass = int(sMassLen/2)
      if(sMassLen%2 == 0):
        print("Колличество букв в слове четное, две средние буквы: ", 
              sMass[sHalfMass - 1],
              sMass[sHalfMass])
      #если_остаток_от_деления_не_равен_0_значит_нечет
      else:
        print("Колличество букв в слове не четное, средняя буква: ",
              sMass[sHalfMass])
    else:
      print("Введенные символы не латинские")

print("\nВремя работы блока:")
```

```python
%%time

#функция_с_тестами_и_проверками
def test_check_simbol():
    #три_теста_для_проверки
    assert check_simbol() != 0 
    assert check_simbol() != 1 
    assert check_simbol() != -1 

print("\nВремя работы блока:")
```

```python
#вызов_функции
test_check_simbol()
```

### Задача_2 (решено)

Вы делаете MVP (минимально жизнеспособный продукт) dating-сервиса. У вас есть список девушек и юношей. Выдвигаем гипотезу: лучшие рекомендации получаются, если просто отсортировать имена по алфавиту и познакомить людей с одинаковыми индексами после сортировки. Но вы не будете никого знакомить, если кто-то может остаться без пары.

```python
#Создание_функции_печати_массивов
def print_mass(massive):
    #цикл_перебирает_элементы_и_выводит_на_печать
    for i in range(len(massive)):
        print(massive[i])
```

```python
#создание_функции_сортировки_массивов
def sorting_mass(massive_1, massive_2):
    #сортировка_массивов
    massive_1.sort()
    massive_2.sort()
```

```python
#создание_функции_ввода_массивов_парней_и_девчонок
def input_massive():
    print("Введите имена парней через пробел:")
    sBoys = input()
    #заполнение_массива_для_парней
    boys = sBoys.split()

    print("\nВведите имена девчонок через пробел:")
    sGirls = input()
    #заполнение_массива_для_девчонок
    girls = sGirls.split()

    return boys, girls 
```

```python
#создание_функции_dating
def dating_service():
    #ввод_исходных_данных
    #boys = ['Peter', 'Alex', 'John', 'Arthtur', 'Richard']
    #girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

    boys, girls = input_massive()

    #проверка_на_равенство_массивов
    if(len(boys) == len(girls)):
       #печать_исходных_массивов_девчонок_и_парней
      print("Исходный массив парней: ")
      print_mass(boys)
      print("\nИсходный массив девчонок: ")
      print_mass(girls)
    
      #сортировка_массивов
      sorting_mass(boys, girls)

      #печать_отсортированного_массива_парней
      print("\nОтсортированный массив парней: ")
      print_mass(boys)
      print("\nОтсортированный массив девчонок: ")
      print_mass(girls) 

      print("\nРезультат:")
      print("\nИдеальные пары:")
      
      #составление_пар
      for i in range(len(boys)):
          print(boys[i],"и",girls[i])
          
    else:
      print("Внимание, кто-то может остаться без пары!")
```

```python
#создание_фукнции_тестов
def dating_test():
    assert dating_service() !=0
```

```python
#вызов_функции_тестов
dating_test()
```
